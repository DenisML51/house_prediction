```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_1.csv')
df.head()
```

```python
def transform_renovation(value):
    if pd.isna(value):
        return np.random.choice([0, 1]) 
    elif value == "Отсутствует":
        return 0
    else:
        return 1

df['renovation'] = df['renovation'].apply(transform_renovation)
```

```python
def transform_deal_type(value):
    if pd.isna(value):
        return np.random.choice([0, 1])  
    elif value == "Свободная продажа":
        return 1
    elif value == "Альтернатива":
        return 0
    else:
        return np.random.choice([0, 1])  

df['deal_type'] = df['deal_type'].apply(transform_deal_type)
```

```python
def transform_balcony(value):
    if pd.isna(value):
        return int(np.clip(np.round(np.random.normal(0.5, 0.2)), 0, 1))
    elif value == "нет":
        return 0
    else:
        return 1

df['balcony'] = df['balcony'].apply(transform_balcony)
```

```python
def transform_trash_chute(value):
    if pd.isna(value) and np.random.rand() < 0.4:
        return np.nan
    elif pd.isna(value):
        return int(np.clip(np.round(np.random.normal(0.5, 0.2)), 0, 1))
    else:
        return 1 

df['trash_chute'] = df['trash_chute'].apply(transform_trash_chute)
```
