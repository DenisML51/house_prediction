```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_EDA.csv')
df.head()
```

```python
def transform_renovation(value):
    if pd.isna(value):
        return np.random.choice([0, 1]) 
    elif value == "Отсутствует":
        return 0
    else:
        return 1

df['Ремонт'] = df['Ремонт'].apply(transform_renovation)
```

```python
def transform_deal_type(value):
    if pd.isna(value):
        return np.random.choice([0, 1])  
    elif value == "Свободная продажа":
        return 1
    elif value == "Альтернатива":
        return 0
    else:
        return np.random.choice([0, 1])  

df['Тип сделки'] = df['Тип сделки'].apply(transform_deal_type)
```

```python
def transform_balcony(value):
    if pd.isna(value):
        return int(np.clip(np.round(np.random.normal(0.5, 0.2)), 0, 1))
    elif value == "нет":
        return 0
    else:
        return 1

df['Балкон'] = df['Балкон'].apply(transform_balcony)
```

```python
def transform_trash_chute(value):
    if pd.isna(value) and np.random.rand() < 0.4:
        return np.nan
    elif pd.isna(value):
        return int(np.clip(np.round(np.random.normal(0.5, 0.2)), 0, 1))
    else:
        return 1 

df['Мусоропровод'] = df['Мусоропровод'].apply(transform_trash_chute)
```

```python
if 'Балкон' in df.columns and 'количество балконов' in df.columns:
    df.loc[(df['Балкон'] == 0) & (df['количество балконов'].notna()) & (df['количество балконов'] != 0), 'Балкон'] = 1
```

```python
if 'Адрес' in df.columns:
    df.drop(columns=['address'], inplace=True)

if 'Unnamed:0.1' in df.columns:
    df.drop(columns=['Unnamed:0.1'], inplace=True)

if 'Unnamed:0' in df.columns:
    df.drop(columns=['Unnamed:0'], inplace=True)
```

```python
columns_translation = {
    'price': 'price',  # Цена
    'adress': 'address',  # Адрес
    'Комнат': 'rooms',  # Количество комнат
    'Площадь': 'area',  # Площадь
    'Кухня': 'kitchen_area',  # Площадь кухни
    'Этаж': 'floor',  # Этаж
    'Ремонт': 'renovation',  # Тип ремонта
    'Тип сделки': 'deal_type',  # Тип сделки
    'Балкон': 'balcony',  # Балкон
    'Мусоропровод': 'trash_chute',  # Мусоропровод
    'Лет в собственности': 'years_in_ownership',  # Лет в собственности
    'Количество собственников': 'number_of_owners',  # Количество собственников
    'Класс энергоэффективности': 'energy_efficiency_class',  # Класс энергоэффективности
    'Энергоснабжение': 'power_supply',  # Энергоснабжение
    'Количество лифтов': 'number_of_elevators',  # Количество лифтов
    'Вентиляция': 'ventilation',  # Вентиляция
    'Грузовой лифт': 'freight_elevator',  # Грузовой лифт
    'Газ': 'gas_supply',  # Газоснабжение
    'latitude': 'latitude',  # Широта
    'longitude': 'longitude'  # Долгота
}

df.rename(columns=columns_translation, inplace=True)
```
